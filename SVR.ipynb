```python
import numpy as np
import pandas as pd
from sklearn.svm import SVR
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('ENERGYMETER2.csv')
```

```python
data.columns = ["Vrms","Irms","PowerFactor","Power","Energy"]
df = pd.DataFrame(data)
df
```

```python
X = []
for i in range(len(df.Energy)) :
    X.append(2.5*i)
X
X = np.array(X)
X = X.reshape(-1,1)
X
```

```python
y = df.Energy
y = np.array(y)
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.6)
```

```python
svr = SVR(kernel='linear')
svr.fit(X_train,y_train)
```

```python
y_pred = svr.predict(X_test)
y_pred
```

```python
r2 = r2_score(y_pred,y_test)
r2
```

```python
rows = X.shape[0]
rows
x = []
x.append(5*X[rows-1])
start = x[0]
```

```python
for i in range(1,20):
    start =  start+3
    x.append(start)
x
```

```python
out = [10*X[rows-1],15*X[rows-1],20*X[rows-1]]
output = []
for i in out :
    output.append(i/60)
output
```

```python
y_predi = svr.predict(out)
y_predi
```

```python
energy_in_units = []
for i in y_predi :
    energy_in_units.append(i/1000)
energy_in_units
```

```python
price = []
for i in energy_in_units :
    price.append(i*3)
price
```
